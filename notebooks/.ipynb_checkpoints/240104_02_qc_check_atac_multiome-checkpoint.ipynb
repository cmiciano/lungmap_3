```r
correction <- "nobatch"
```

```r
# Name the reductions based on whether or not harmony was applied
red_nm_rna <- ifelse(correction == "batch", "harmony.rna", "pca")
red_nm_rna_umap <- ifelse(correction == "batch", "umap.hrm.rna", "umap.rna")
red_nm_atac <- ifelse(correction == "batch", "harmony.atac", "lsi")
red_nm_atac_umap <- ifelse(correction == "batch", "umap.hrm.atac", "umap.atac")
red_nm_wnn_umap <- ifelse(correction == "batch", "umap.hrm.wnn", "umap.wnn")


red_nm_rna 
red_nm_rna_umap 
red_nm_atac
red_nm_atac_umap 
red_nm_wnn_umap
```

```r
library(dplyr)
```

```r
## ATAC processing
DefaultAssay(sobj) <- 'ATAC'

print("Processing ATAC")
Sys.time()
sobj <- RunTFIDF(sobj) %>%
FindTopFeatures(min.cutoff = 'q0') %>%
RunSVD()
Sys.time()

if (correction == "batch") {
    sobj <- RunHarmony(sobj, reduction = "lsi", group.by.vars = 'orig.ident',
               assay.use = DefaultAssay(sobj),
               reduction.save = red_nm_atac, project.dim = F)
}

Sys.time()
sobj <- RunUMAP(sobj, reduction = red_nm_atac , dims = 2:30, verbose = F,
           reduction.name=red_nm_atac_umap, reduction.key='atac_UMAP_') %>%
        FindNeighbors(reduction = red_nm_atac , dims = 2:30) %>%
        FindClusters(algorithm = 3, resolution = .3, verbose = FALSE)
Sys.time()

# UMAPS by library and seurat clusters
# NOTE
pdf(paste0(outdir,  gsub("\\.", "_", red_nm_atac_umap), "_" , correction, "_libid.pdf"))
DimPlot(sobj, reduction = red_nm_atac_umap, group.by = "orig.ident")
dev.off()


pdf(paste0(outdir,  gsub("\\.", "_", red_nm_atac_umap),  "_" , correction, "_clust.pdf"))
DimPlot(sobj, reduction = red_nm_atac_umap, group.by = "seurat_clusters")
dev.off()
```

```r
DimPlot(sobj, reduction = red_nm_atac_umap, group.by = "seurat_clusters")
```

```r
### WNN processing
print("Running WNN")
Sys.time()
sobj <- FindMultiModalNeighbors(sobj, reduction.list=list(red_nm_rna, red_nm_atac),
            dims.list=list(1:50, 2:30),
            weighted.nn.name = "weighted.nn",
            snn.graph.name = "wsnn")
Sys.time()


Sys.time()
sobj <- RunUMAP(sobj, nn.name='weighted.nn', reduction.name=red_nm_wnn_umap, reduction.key='wnn_UMAP_',
verbose = F)
Sys.time()


# try algorithm 3 instead of 4 and method 'igraph'
sobj <- FindClusters(sobj, graph.name='wsnn', algorithm=3,  resolution = 0.25, verbose=FALSE)

# NOTE
#pdf(paste0(outdir,  gsub("\\.", "_", red_nm_wnn_umap), "_" , correction, "_libid.pdf"))
DimPlot(sobj, reduction = red_nm_wnn_umap, group.by = "orig.ident")
dev.off()

#pdf(paste0(outdir,  gsub("\\.", "_", red_nm_wnn_umap), "_" , correction, "_clust.pdf"))
DimPlot(sobj, reduction = red_nm_wnn_umap, group.by = "seurat_clusters")
dev.off()

options(repr.plot.width=18, repr.plot.height=6)
p1 <- DimPlot(sobj , reduction=red_nm_rna_umap, group.by='seurat_clusters', label=TRUE, label.size=3, repel=TRUE) + ggtitle('RNA')
p1 <- p1 + xlab('UMAP 1') + ylab('UMAP 2') + ggtitle('RNA only')
p2 <- DimPlot(sobj , reduction=red_nm_atac_umap, group.by='seurat_clusters', label=TRUE, label.size=3, repel=TRUE) + ggtitle('ATAC')
p2 <- p2 + xlab('UMAP 1') + ylab('UMAP 2') + ggtitle('ATAC only')
p3 <- DimPlot(sobj , reduction=red_nm_wnn_umap, group.by='seurat_clusters', label=TRUE, label.size=3, repel=TRUE) + ggtitle('WNN')
p3 <- p3 + xlab('UMAP 1') + ylab('UMAP 2') + ggtitle('WNN - Combined')
p4 <- p1 + p2 + p3 & NoLegend()

```
