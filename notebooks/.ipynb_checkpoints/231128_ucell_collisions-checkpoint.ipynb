```r
# runs ucell on particular list
# outputs featureplots 
# outputs boxplots of senescent vs non sen
# outputs boxplots of condition

# ask whether to plot pdfs or not
# ask what columns to plot

# gene_list: input gene list as a vector
# gl_col_name: name of metadata column and how it'll be displayed
# outdir: path of ucell pdfs to be written
# by default plot condition
run_ucell <- function(seu_obj, gene_list, gl_col_name, outdir ) {
    gene_list <- list(gene_list_vec)
    names(gene_list) <- gl_col_name
    
    print("Adding UCell scores")
    seu_obj <- AddModuleScore_UCell(seu_obj, features = gene_list)
    
    dir.create(paste0(outdir, "ucell/"))
    dir.create(paste0(outdir, "ucell/pdfs/"))
    pdf_out <- paste0(outdir, "ucell/pdfs/")
    
    print("Writing UCell featureplots")
    pdf(paste0(pdf_out, gl_col_name, "_UCell", "_fp.pdf"))
    plt_fp <- FeaturePlot(seu_obj, features = paste0(gl_col_name, "_UCell")) + scale_color_viridis(option = "plasma", direction = -1)
    print(plt_fp)
    dev.off()
    
    
    #print("Writing UCell boxplots")
    #pdf(paste0(pdf_out, gl_col_name, "_UCell", "_senescent", "_bp.pdf"))
    #plt_bp <- dittoBoxPlot(seu_obj,  paste0(gl_col_name, "_UCell"), group.by = "senescent", legend.show = FALSE) +
    #                theme(text = element_text(size = 20), plot.margin = margin(10,10,10,10)) + 
    #                geom_boxplot(outlier.shape=NA) + 
    #                labs( x="senescent", y="UCell Enrichment Score")
    #print(plt_bp)
    #dev.off()
    
    
    
    #pdf(paste0(pdf_out, gl_col_name, "_UCell", "_condition", "_bp.pdf"))
    #plt_rp <- dittoBoxPlot(seu_obj,  paste0(gl_col_name, "_UCell"), group.by = "condition", legend.show = FALSE) +
    #                theme(text = element_text(size = 20), plot.margin = margin(10,10,10,10)) + 
    #                geom_boxplot(outlier.shape=NA) + 
    #               labs( x="condition", y="UCell Enrichment Score") 
    #print(plt_rp)
    #dev.off()

    print("Writing UCell Table")

    ucell_col <- grep(gl_col_name, colnames(seu_obj@meta.data), value = T)

    ucell_df <- data.frame(seu_obj[[ucell_col]])
    dir.create(paste0(outdir, "ucell/table/"))
    tb_out <- paste0(outdir, "ucell/table/")
    write.table(ucell_df, paste0(tb_out, gl_col_name, "_ucell.txt"))
    return(seu_obj)
}
```
